```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df =pd.read_csv(r"C:\Users\USER\Downloads\world_population.csv")
df
```

```python
df.sort_values(by='2022 Population',ascending=False).head(10)
```

```python
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['number']) 
result = df.groupby('Continent')[numeric_cols.columns].mean().sort_values(by='2022 Population',ascending=False)

# Display the result
print(result)
```

```python
result.plot()
```
